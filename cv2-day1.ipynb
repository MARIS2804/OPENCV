```python
import cv2
import matplotlib.pylab as plt
```

```python
img='bb.jpg'
img_cv=cv2.imread(img)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_cv,cmap="Reds")
ax.axis("off")
plt.show()
```

```python
fig,ax=plt.subplots(1,3,figsize=(7,7))
ax[0].imshow(img_cv[:,:,0],cmap="Reds")
ax[1].imshow(img_cv[:,:,1],cmap="Blues")
ax[2].imshow(img_cv[:,:,2],cmap="Greens")
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
plt.show()
```

```python
img_gray=cv2.cvtColor(img_cv,cv2.COLOR_RGB2GRAY)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_gray,cmap="Greys")
ax.axis("off")
plt.show()
```

```python
img_resize1=cv2.resize(img_cv,None,fx=0.5,fy=0.5)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_resize1)
ax.axis("off")
plt.show()
```

```python
img_resize2=cv2.resize(img_cv,(50,50))
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_resize2)
ax.axis("off")
plt.show()
```

```python
img_resize3=cv2.resize(img_cv,(50,50),interpolation=cv2.INTER_CUBIC)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_resize3)
ax.axis("off")
plt.show()
```

```python
cv2.MORPH_OPEN
```

```python
#sharpening
import numpy as np
kernel=np.array([[-1,-1,-1],
                [-1,9,-1],
                [-1,-1,-1]])
img_sharp=cv2.filter2D(img_cv,-1,kernel)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_sharp)
ax.axis("off")
plt.show()
```

```python
import numpy as np
kernel=np.array([[-1,-1,-1],
                [-1,8,-1],
                [-1,-1,-1]])
img_edge=cv2.filter2D(img_cv,-1,kernel)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_edge)
ax.axis("off")
plt.show()
```

```python
import numpy as np
kernel=np.array([[1,1,1],
                [1,1,1],
                [1,1,1]])
kernel=kernel/9
img_blur=cv2.filter2D(img_cv,-1,kernel)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(img_blur)
ax.axis("off")
plt.show()
```

```python
kernel=np.ones((3,3),np.uint8)
erode=cv2.erode(img_cv,kernel,iterations=1)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(erode)
ax.axis("off")
plt.show()
```

```python
kernel=np.ones((3,3),np.uint8)
dilate=cv2.dilate(img_cv,kernel,iterations=1)
fig,ax=plt.subplots(figsize=(4,4))
ax.imshow(dilate)
ax.axis("off")
plt.show()
```
